```python
imgs = []

for file in tqdm(os.listdir(f'{PATH}/valid/cats')):
    img = plt.imread(f'{PATH}/valid/cats/{file}')
    imgs.append(img.shape)
    
for file in tqdm(os.listdir(f'{PATH}/valid/dogs')):
    img = plt.imread(f'{PATH}/valid/dogs/{file}')
    imgs.append(img.shape)
```

```python
hist_w, bins_w = np.histogram(np.array(imgs)[:,0])
x = (bins_w[:-1] + bins_w[1:])/2
y = hist_w
plt.plot(x, y)

hist_h, bins_h = np.histogram(np.array(imgs)[:,1])
x = (bins_h[:-1] + bins_h[1:])/2
y = hist_h
plt.plot(x, y)
plt.show()
```
